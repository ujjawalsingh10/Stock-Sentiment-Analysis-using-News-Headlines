```python
import pandas as pd
```

```python
df = pd.read_csv('Data.csv', encoding = 'ISO-8859-1')
```

```python
df
```

```python
df['Date'].value_counts()
```

```python
X = df.drop(['Label', 'Date'], axis = 1)
y = df['Label'].values
```

```python
#Removing punctuations
X.replace("[^a-zA-Z]", " ", regex = True, inplace = True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y , test_size = 0.2, random_state = 1)
```

```python
#Renmaing column names for ease of access
list1 = [i for i in range(25)]
new_index = [str(i) for i in list1]
X_train.columns = new_index
X_train.head(5)
```

```python
#Making the headlines lower cased
for index in new_index:
    X_train[index] = X_train[index].str.lower()
X_train.head(2)
```

```python
' '.join(str(x) for x in X_train.iloc[0,:])
```

```python
headlines = []
for row in range(0, len(X_train.index)):
    headlines.append(' '.join(str(x) for x in X_train.iloc[row, :]))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
#Implement bag of words
countVector = CountVectorizer(ngram_range = (2,2))
trainDataset = countVector.fit_transform(headlines)
```

```python
#implementing random forest classifier
rfc = RandomForestClassifier(200, criterion = 'entropy')
rfc.fit(trainDataset, y_train)
```

```python
#Predict for the test dataset
test_transform= []
for row in range(len(X_test.index)):
    test_transform.append(' '.join(str(x) for x in X_test.iloc[row,:]))
test_dataset = countVector.transform(test_transform)
y_pred = rfc.predict(test_dataset)
```

```python
#Check accuracy
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
matrix = confusion_matrix(y_test, y_pred)
print(matrix)
score = accuracy_score(y_test, y_pred)
print(score)
report = classification_report(y_test, y_pred)
print(report)
```
